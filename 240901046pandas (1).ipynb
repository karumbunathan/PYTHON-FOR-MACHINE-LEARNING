```python
import pandas as pd
#empty dataframe
df=pd.DataFrame()
print(df)
```

```python
emp=pd.Series(['Parker','John','smith','William'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}
result=pd.DataFrame(frame)
print("\nSeries to data frame\n")
print(result)
```

```python
#adding new column
print("\n adding new column:\n")
result['Age']=pd.Series([35,24,40,38])
print(result)
```

```python
#deleting column
print("\n Deleting one colunm:\n")
del result['Age']
print(result)
```

```python
#extracting 3rd row
print("\n Extracting third row:\n")
print(result.loc[2])
```

```python
#slicing
print("\n slice rows:\n",result[1:3])
```

```python
d2=pd.DataFrame([['Dale',123],['Mark',143]],columns=['Emp','ID'])
result=pd.concat([result,d2],ignore_index=True)
print("\nAdding new row values:\n",result)
```

```python
print("\nDeleting particular row:\n",result.drop(1))
```
