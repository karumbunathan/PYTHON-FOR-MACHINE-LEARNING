```python
import pandas as pd
pd.set_option('display.max_columns',10)
print("Original data:\n")
df=pd.read_excel("data cleaning.xlsx")
print(df)
```

```python
f=df['NAME'].str.strip()
f.to_excel("data cleaning.xlsx")
print("\nReplace value:")
k=df.fillna('pad')
print(k)
print("\nUpdated list\n")
k.to_excel("data cleaning.xlsx")
```

```python
#Updated list
print("\nDrop one particular column and its values:\n")
k.drop(['SOCIAL'], axis=1,inplace=True)
print(k)
print("\nUpdated list..\n")
k.to_excel("data cleaning.xlsx")
```

```python
#Updated list..
print("\nDrop NaN rows:\n")
df=pd.read_excel("data cleaning.xlsx")
print(df)
x=df.dropna()
print(x)
print("\nUpdated data list:\n")
x.to_excel("data cleaning.xlsx")
```
